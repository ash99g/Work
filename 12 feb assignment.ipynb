```python
# Q.5 What are Custom exceptions in python? why do we need custom exception? explain with an example.
# Ans. 1. In Python, we can define custom exceptions by creating a new class that is derived from the built-in Exception class.
#      2.To catch and provide specific treatment to a subset of existing exceptions.
#        Built-in exceptions offer information about Python-related problems, and 
#        custom exceptions will add information about project-related problems.


class SalaryNotInRangeError(Exception):

    def __init__(self, message):
        self.message = message
```

```python
def salaryrange(salary):
    if salary < 20000 :
        raise SalaryNotInRangeError("very low salary")
    elif salary <=20000 and salary >= 50000 :
        raise SalaryNotInRangeError("medium salary")
    elif salary > 80000 :
        raise SalaryNotInRangeError(" high salary")
    else :
        print("good salary")
```

```python
try :
    salary = int(input("enter your salary"))
    salaryrange(salary)
except SalaryNotInRangeError as e :
    print(e)
```
