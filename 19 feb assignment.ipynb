```python
# Ans.To build a URL to a specific function, use the url_for() function. It accepts the name of the function as its first argument and any number of keyword arguments, each corresponding to a variable part of the URL rule.
```

```python
    from flask import *  
       
    app = Flask(__name__)  
      
    @app.route('/admin')  
    def admin():  
        return 'admin'  
      
    @app.route('/librarion')  
    def librarion():  
        return 'librarion'  
      
    @app.route('/student')  
    def student():  
        return 'student'  
      
    @app.route('/user/<name>')  
    def user(name):  
        if name == 'admin':  
            return redirect(url_for('admin'))  
        if name == 'librarion':  
            return redirect(url_for('librarion'))  
        if name == 'student':  
            return redirect(url_for('student'))  
    if __name__ =='__main__':  
        app.run(debug = True)  
```
